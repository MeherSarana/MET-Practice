```python
# Simple Calci

a = int(input("Enter First Number: "))
b = int(input("Enter Second Number: "))
print(f"The addition of {a} & {b} is : {a+b}")
print(f"The Subtraction of {a} & {b} is : {a-b}")
print(f"The Multiplication of {a} & {b} is : {a*b}")
print(f"The Division of {a} & {b} is : {a/b}")
print(f"The Power of {a} & {b} is : {a**b}  {b**a}")
```

```python
# Ask user to enter credit Card Number
# Print the Number with last four digits masked with XXXX

credit_no = input("Enter credit Card Number")
output = "XXXX XXXX XXXX YYYY"

last_digits = credit_no[-4::]
final_word = output.replace("YYYY", last_digits)
print(final_word)
```
