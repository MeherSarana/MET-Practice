```python
# Logical Operators - and, or, not

print(False or False)
print(False or True)
print(True or True)
print(True or False)


print(False and False)
print(False and True)
print(True and True)
print(True and False)
```

```python
# Identity Operators - is, is not (Based on addressing and the similar values in the loc) 

x = 10
y = 10

print(x is y)
print(x is not y)
```

```python
# Comparison Operators - ==, !=, >, <, >=, <=

a= 10
b= 15

print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)
```
