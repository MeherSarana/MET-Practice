```python
questions = [
{
"question": "1: What is the Output Of 2+2*3 ? ",
"options": ["A. 6", "B. 7", "C. 8", "D. 9"],
"answer": "C"
},
{
"question": "2: What is the Output Of 15//4 ? ",
"options": ["A. 3", "B. 3.5", "C. 6", "D. 4"],
"answer": "A"
},
{
"question": "2: What is the Output Of 2**3 ? ",
"options": ["A. 6", "B. 8", "C. 4", "D. 10"],
"answer": "B"
},
]

score = 0
attempt = []
for i,q in enumerate(questions,1):
    print(f"Question {i}: {q['question']}")
    for option in q["options"]:
        print(option)
    print(" ")

    answer = input("Enter the Options (A/B/C/D):")
    attempt_entry = (i,answer,q["answer"])
    if answer == q["answer"]:
        score+=2
    else:
        score-=0.5
    attempt.append(attempt_entry)
    
print("Final Score:",score,"Out of ",len(questions))   

print("No. **** Your Answer **** Correct Answer")
for quest in attempt:
    index_no, your_ans, correct_ans = quest
    print(f"{index_no} \t\t {your_ans} \t\t {correct_ans}")
```
