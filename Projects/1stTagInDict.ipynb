```python
import requests
import streamlit as st
 
print("1st Tag in the Dictionary")
choice = input('Enter a word:')
endpoint = f"https://api.datamuse.com/words?ml=i%20am%20{choice}"
 
response = requests.get(endpoint)
 
data = response.json()
 
if response.status_code == 200:
    for item in data:
        tags = item.get('tags')
        print(tags[0])
```
