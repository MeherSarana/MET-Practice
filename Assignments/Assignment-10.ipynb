```python
#1
class Device:
    def __init__(self,brand,color,connection):
        self.brand = brand
        self.color = color
        self.connection = connection
    
    def turn_on(self):
        print(f"{self.brand} is Switched On")

    def reset(self):
        print(f"{self.brand} has been Reset")

dev1 = Device("Samsung","Blue","Wired")
dev2 = Device("TP-Link Router","White","Wireless")
dev3 = Device("LG Moniter","Cyan","Wired")

print(dev1.turn_on())
print(dev2.brand)
print(dev3.reset())        
```

```python
#2
class Car:

    def __init__(self,model,speed, fuel_type):
        self.model =model
        self.speed = speed
        self.fuel_type = fuel_type

    def accelerate(self):
        print(f"The Car is Accelerating at Speed {self.speed}")

    def stop(self):
        print(f"The Car has been Stopped")
        
    def honk(self):
        print(f"Pee Pee")

class Bike:

    def __init__(self,model,speed, fuel_type):
        self.model =model
        self.speed = speed
        self.fuel_type = fuel_type

    def accelerate(self):
        print(f"The Bike is Accelerating at Speed {self.speed}")

    def stop(self):
        print(f"The Car has been Stopped")

    def honk(self):
        print(f"Pee Pee")

class Truck:

    def __init__(self,model,speed, fuel_type):
        self.model =model
        self.speed = speed
        self.fuel_type = fuel_type

    def accelerate(self):
        print(f"The Truck is Accelerating at Speed {self.speed}")

    def stop(self):
        print(f"The Car has been Stopped")

    def honk(self):
        print(f"Pee Pee")

car = Car("M340i","90","Petrol")
bike = Bike("S1000RR","160","Petrol")
truck = Truck("Volvo","40","Disel")

car.stop()
bike.accelerate()
truck.honk()
```

```python
#3
class Student:
    def __init__(self, name, age, student_id):
        self.name = name
        self.age = age
        self.student_id = student_id

    def introduce(self):
        print(f"I am {self.name}. My ID is {self.student_id}.")

class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def teach(self):
        print(f"{self.name} is teaching {self.subject}.")

class Classroom:
    def __init__(self, room_number, teacher, students):
        self.room_number = room_number
        self.teacher = teacher
        self.students = students  

    def class_info(self):
        print(f"Classroom {self.room_number} is taught by {self.teacher.name}.")
        print("Students in this classroom:")
        for student in self.students:
            print(f"- {student.name}")

student1 = Student("Alice", 14, "S101")
student2 = Student("Bob", 15, "S102")
teacher1 = Teacher("Mr. Smith", "Mathematics")
classroom1 = Classroom("A1", teacher1, [student1, student2])

student1.introduce()
teacher1.teach()
classroom1.class_info()   
```

```python
#4
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def view_details(self):
        print(f"Product: {self.name}, Price: {self.price}, Stock: {self.stock}")

    def update_stock(self, quantity):
        self.stock += quantity
        print(f"Stock for {self.name} updated to {self.stock}")

class Cart:
    def __init__(self):
        self.items = []

    def add_product(self, product):
        if product.stock > 0:
            self.items.append(product)
            product.stock -= 1
            print(f"{product.name} added to cart.")
        else:
            print(f"{product.name} is out of stock.")

    def view_cart(self):
        print("Cart contains:")
        for item in self.items:
            print(f"- {item.name}")

class User:
    def __init__(self, username, email, address):
        self.username = username
        self.email = email
        self.address = address
        self.cart = Cart()

    def add_to_cart(self, product):
        self.cart.add_product(product)

    def view_profile(self):
        print(f"User: {self.username}, Email: {self.email}, Address: {self.address}")

product1 = Product("Laptop", 50000, 5)
product2 = Product("Headphones", 2000, 2)

user1 = User("john_doe", "john@example.com", "123 Main St")

product1.view_details()
user1.view_profile()
user1.add_to_cart(product1)
user1.add_to_cart(product2)
user1.cart.view_cart()
product1.update_stock(10)
```

```python
#5
class Book:
    def __init__(self, title, author, ID):
        self.title = title
        self.author = author
        self.id = ID
        self.issued = False

    def read(self):
        print(f"Reading '{self.title}' by {self.author}.")

    def issue(self):
        if not self.issued:
            self.issued = True
            print(f"'{self.title}' has been issued.")
        else:
            print(f"'{self.title}' is already issued.")

    def return_book(self):
        if self.issued:
            self.issued = False
            print(f"'{self.title}' has been returned.")
        else:
            print(f"'{self.title}' was not issued.")

class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.books_issued = []

    def issue_book(self, book):
        if not book.issued:
            book.issue()
            self.books_issued.append(book)
            print(f"{self.name} has issued '{book.title}'.")
        else:
            print(f"{book.title} is not available.")

    def return_book(self, book):
        if book in self.books_issued:
            book.return_book()
            self.books_issued.remove(book)
            print(f"{self.name} has returned '{book.title}'.")
        else:
            print(f"{self.name} does not have '{book.title}'.")

class Librarian:
    def __init__(self, name):
        self.name = name

    def add_book(self, book, library):
        library.append(book)
        print(f"Librarian {self.name} added '{book.title}' to the library.")

    def remove_book(self, book, library):
        if book in library:
            library.remove(book)
            print(f"Librarian {self.name} removed '{book.title}' from the library.")
        else:
            print(f"'{book.title}' not found in the library.")

library = []
book1 = Book("1984", "George Orwell", "12345")
book2 = Book("To Kill a Mockingbird", "Harper Lee", "67890")
librarian = Librarian("Ms. Green")
member = Member("Alice", "M001")

librarian.add_book(book1, library)
librarian.add_book(book2, library)
# print(library[:])
member.issue_book(book1)
book1.read()
member.return_book(book1)
librarian.remove_book(book2,library)
```
